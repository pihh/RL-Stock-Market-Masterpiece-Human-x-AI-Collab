```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import RobustScaler
# === Step 1: Load Stock Data ===
#data = yf.download('AAPL', start='2020-01-01', end='2024-12-31')
#data = data[['Close']]
#data['Returns'] = np.log(data['Close'] / data['Close'].shift(1))
#data.dropna(inplace=True)
data = ohlcv_df[ohlcv_df['symbol']=="AAPL"]
data.set_index('date',inplace=True)
data = data[['close','volume']]
data.rename(columns={"close":"Close"}, inplace=True)
data['Returns']=np.log(data['Close'] / data['Close'].shift(1))
data.dropna(inplace=True)


returns = data['Returns'].values
mean = returns.mean()
std = returns.std()
#normalized_returns = (returns - mean) / std

volumes = data['volume'].values
mean_v = volumes.mean()
std_v = volumes.std()
#normalized_volumes = (volumes - mean_v) / std_v

pdf = pd.DataFrame({"volumes":returns,"returns":volumes})
scaler = RobustScaler()
scaled = scaler.fit(pdf)
ds = scaled.transform(pdf)
#print(dataset)
seq_len = 20
X, y = [], []
for i in range(len(normalized_returns) - seq_len):
    #X.append(normalized_returns[i:i+seq_len])
    X.append(ds[i:i+seq_len])
    y.append(ds[i+seq_len][1])
X, y = np.array(X), np.array(y)

import pennylane as qml
from pennylane import numpy as np

n_qubits = 4
dev = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev, interface="torch")
def quantum_circuit(inputs, weights):
    for i in range(n_qubits):
        qml.RY(inputs[i], wires=i)
    qml.templates.BasicEntanglerLayers(weights, wires=range(n_qubits))
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

# === Step 2: Define Quantum-Inspired RNN Cell ===
class QuantumRNNCell(nn.Module):
    def __init__(self, input_size, hidden_size):
        super().__init__()
        self.hidden_size = hidden_size
        self.Wr = nn.Parameter(torch.randn(input_size, hidden_size) * 0.1)
        self.Wi = nn.Parameter(torch.randn(input_size, hidden_size) * 0.1)
        self.Ur = nn.Parameter(torch.randn(hidden_size, hidden_size) * 0.1)
        self.Ui = nn.Parameter(torch.randn(hidden_size, hidden_size) * 0.1)
        self.activation = torch.tanh

    def forward(self, x, h_r, h_i):
        x = x.unsqueeze(1)
        x_r = x @ self.Wr
        x_i = x @ self.Wi
        h_r_new = self.activation(x_r + h_r @ self.Ur - h_i @ self.Ui)
        h_i_new = self.activation(x_i + h_r @ self.Ui + h_i @ self.Ur)
        return h_r_new, h_i_new

# === Step 3: Define Full QRNN Model ===
class QRNN(nn.Module):
    def __init__(self, input_size, hidden_size):
        super().__init__()
        self.cell = QuantumRNNCell(input_size, hidden_size)
        self.hidden_size = hidden_size
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        #print(x.shape)
        batch_size, seq_len ,_= x.shape
        h_r = torch.zeros(batch_size, self.hidden_size)
        h_i = torch.zeros(batch_size, self.hidden_size)
        for t in range(seq_len):
            h_r, h_i = self.cell(x[:, t], h_r, h_i)
        output = self.fc(h_r)
        return output.squeeze()

# === Step 4: Prepare Data for PyTorch ===
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32)
dataset = TensorDataset(X_tensor, y_tensor)
loader = DataLoader(dataset, batch_size=32, shuffle=True)

# === Step 5: Train the Model ===
model = QRNN(input_size=2, hidden_size=32)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_fn = nn.MSELoss()

for epoch in range(20):
    total_loss = 0
    for xb, yb in loader:
        pred = model(xb)
        loss = loss_fn(pred, yb)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}: Loss = {total_loss / len(loader):.6f}")

# === Step 6: Evaluate ===
with torch.no_grad():
    preds = model(X_tensor).numpy()

plt.figure(figsize=(12, 4))
plt.plot(y, label='True')
plt.plot(preds, label='Predicted')
plt.title("True vs Predicted Log-Returns")
plt.legend()
plt.grid()
plt.show()
```
